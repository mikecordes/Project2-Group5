```python
#dependencies and libraries
import numpy as np
import pandas as pd
import hvplot.pandas
from sklearn.preprocessing import MinMaxScaler
from numpy.random import seed
from tensorflow import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import glob
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
root_directory = r"Resources"

all_files = glob.glob(f'{root_directory}\*.csv')

li = []

for filename in all_files:
    frame = pd.read_csv(filename, index_col=None, header=0,)
    frame["Year"] = filename[10:14]
    li.append(frame)

df = pd.concat(li, axis=0, ignore_index=True)
```

# Linear Regression Test For Quarterbacks: PassingTDs and FantasyScore

```python
qb_plot = qbdf.hvplot.scatter(
    x="PassingTD",
    y="FantasyPoints",
    title="Fantasy points per passing Tds"
)
qb_plot
```

```python
X = qbdf["PassingTD"].values.reshape(-1, 1)
X[:5]
```

```python
y = qbdf["FantasyPoints"].values.reshape(-1, 1)
y[:5]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
print(f"Model's slope: {model.coef_}")
```

```python
print(f"Model's y-intercept: {model.intercept_}")
```

```python
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]}X")
```

```python
predicted_y_values = model.predict(X)
```

```python
qbdf_predicted_fantasy_points = qbdf.copy()
```

```python
qbdf_predicted_fantasy_points["PredictedFantasyPoints"]=predicted_y_values
qbdf_predicted_fantasy_points.head()
```

```python
qb_best_fit_line = qbdf_predicted_fantasy_points.hvplot.line(
    x = "PassingTD",
    y = "PredictedFantasyPoints",
    color = "red"
)
qb_best_fit_line
```

```python
qb_plot * qb_best_fit_line
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
score = model.score(X, y, sample_weight=None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)

# Print releveant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
